Q21. What are the assumptions for applying the Central Limit Theorem?
- Answer The Central Limit Theorem (CLT) is one of the most important concepts in statistics because it allows us to make inferences about a population even when we do not know the population distribution. However, the CLT has certain assumptions that must be met for it to apply correctly. These assumptions are as follows:

1. Random Sampling
Assumption: The data should be obtained through random sampling. This ensures that each individual observation in the sample has an equal chance of being selected, which minimizes bias and makes the sample representative of the population.

Reason: Random sampling ensures that the sample is representative of the population, and it helps in generalizing the sample results to the broader population.

2. Independence of Observations
Assumption: The observations in the sample must be independent of one another. This means that the outcome of one observation should not influence the outcome of another observation.

Reason: The independence assumption is important because it ensures that the Central Limit Theorem applies. If observations are not independent (e.g., in a time-series study or when there are strong correlations between observations), the sample mean distribution may not behave as a normal distribution.

3. Sample Size
Assumption: The sample size must be sufficiently large for the CLT to hold. While there is no strict rule, a common rule of thumb is that the sample size should be greater than 30 for most situations. For highly skewed populations or distributions with outliers, a larger sample size may be needed to approximate normality.

Reason: The CLT states that as the sample size increases, the sampling distribution of the sample mean approaches a normal distribution. If the sample size is small, the approximation to normality may not hold, especially if the population distribution is highly skewed or has extreme outliers.

4. Population Distribution
Assumption: The population from which the sample is drawn can have any distribution (it doesn't need to be normal). The Central Limit Theorem guarantees that, regardless of the shape of the population distribution, the distribution of the sample mean will approximate a normal distribution as the sample size becomes large enough.

Reason: If the population distribution is heavily skewed or has extreme outliers, the sample mean distribution will converge to normal more slowly. In such cases, a larger sample size (often greater than 30) is necessary to ensure that the distribution of the sample mean is approximately normal.

5. Finite Variance
Assumption: The population from which the sample is drawn must have a finite variance. This means the population variance (or standard deviation) must be well-defined and not infinite.

Reason: If the population has an infinite variance (e.g., a Cauchy distribution), the Central Limit Theorem does not apply because the sample mean will not converge to a normal distribution, regardless of the sample size.

6. No Extreme Outliers (for Small Sample Sizes)
Assumption: For small sample sizes (typically less than 30), the population should not have extreme outliers. Outliers can heavily skew the distribution, and with small sample sizes, the CLT may not hold as expected.

Reason: Outliers can distort the sample mean significantly, and with a small sample size, this distortion can prevent the sample mean from approaching a normal distribution. In large samples, however, the impact of outliers is often minimized, and the CLT holds more reliably.

```python
## Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution
def bernoulli_pdf(x, p):
    """
    Calculate the probability mass function (PMF) for a Bernoulli distribution.

    Parameters:
    x (int): The outcome of the random variable (0 or 1).
    p (float): The probability of success (0 <= p <= 1).

    Returns:
    float: The probability of the outcome x.
    """
    if x not in [0, 1]:
        raise ValueError("x must be either 0 or 1.")

    # PDF for Bernoulli distribution
    if x == 1:
        return p
    elif x == 0:
        return 1 - p

# Example usage
p = 0.7  # Probability of success (e.g., the probability of getting a 1)
x = 1     # Outcome (success)
print(f"P(X = {x}) = {bernoulli_pdf(x, p)}")

x = 0     # Outcome (failure)
print(f"P(X = {x}) = {bernoulli_pdf(x, p)}")
```

```python
##  Write a Python function to calculate Z-scores from a dataset and plot them.
import numpy as np
import matplotlib.pyplot as plt

# Function to calculate Z-scores and plot them
def calculate_and_plot_z_scores(data):
    """
    Calculate Z-scores for a dataset and plot them.

    Parameters:
    data (array-like): The dataset for which to calculate the Z-scores.
    """
    # Calculate mean (mu) and standard deviation (sigma)
    mean = np.mean(data)
    std_dev = np.std(data)

    # Calculate Z-scores for each data point
    z_scores = (data - mean) / std_dev

    # Plot the Z-scores
    plt.figure(figsize=(10, 6))
    plt.plot(z_scores, marker='o', linestyle='-', color='b', label='Z-scores')

    # Add labels and title
    plt.xlabel('Index')
    plt.ylabel('Z-score')
    plt.title('Z-scores of the Dataset')
    plt.axhline(y=0, color='r', linestyle='--')  # Add a line at Z = 0 (mean)
    plt.legend()
    plt.grid(True)

    # Show the plot
    plt.show()

    # Return the Z-scores for potential further use
    return z_scores

# Example usage
data = np.random.normal(0, 1, 1000)  # Generate a random dataset from a normal distribution
z_scores = calculate_and_plot_z_scores(data)
```

```python
## Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal disribution.
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def calculate_z_score(x, mean, std_dev):
    """
    Calculate the Z-score for a given data point x.

    Parameters:
    x (float): The data point for which the Z-score is calculated.
    mean (float): The mean of the distribution.
    std_dev (float): The standard deviation of the distribution.

    Returns:
    float: The Z-score of the data point.
    """
    return (x - mean) / std_dev

def compare_to_standard_normal(z_score):
    """
    Compare the Z-score to the standard normal distribution.

    Parameters:
    z_score (float): The calculated Z-score.

    Returns:
    float: The cumulative probability (p-value) for the given Z-score in the standard normal distribution.
    """
    # Cumulative probability (CDF) for the standard normal distribution
    return stats.norm.cdf(z_score)

# Example data
data = [12, 15, 19, 23, 21, 17, 14, 18, 20, 16]  # Example dataset

# Step 1: Calculate the mean and standard deviation of the data
mean = np.mean(data)
std_dev = np.std(data)

# Step 2: Calculate the Z-score for a specific data point (e.g., x = 19)
x = 19
z_score = calculate_z_score(x, mean, std_dev)

# Step 3: Compare Z-score to the standard normal distribution
p_value = compare_to_standard_normal(z_score)

# Step 4: Display results
print(f"Data: {data}")
print(f"Mean: {mean}, Standard Deviation: {std_dev}")
print(f"Z-score for data point {x}: {z_score}")
print(f"Cumulative probability (p-value) for Z-score: {p_value}")

# Step 5: Visualize the standard normal distribution and the Z-score
# Generate values for the standard normal distribution
x_vals = np.linspace(-4, 4, 1000)
y_vals = stats.norm.pdf(x_vals, loc=0, scale=1)

# Plot the standard normal distribution
plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label="Standard Normal Distribution", color="blue")

# Plot the Z-score on the normal distribution curve
plt.axvline(x=z_score, color="red", linestyle="--", label=f"Z-score: {z_score:.2f}")

# Labels and title
plt.title("Standard Normal Distribution and Z-score Comparison")
plt.xlabel("Z")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

```python
##  Implement hypothesis testing using Z-statistics for a sample dataset5
import numpy as np
import scipy.stats as stats

def z_test(sample_data, population_mean, population_std, alpha=0.05):
    """
    Perform a one-sample Z-test to test if the sample mean is significantly different
    from the population mean.

    Parameters:
    sample_data (array-like): The sample dataset.
    population_mean (float): The population mean under the null hypothesis.
    population_std (float): The population standard deviation.
    alpha (float): The significance level (default is 0.05).

    Returns:
    None: Prints the test results.
    """
    # Step 1: Calculate the sample mean and sample size
    sample_mean = np.mean(sample_data)
    n = len(sample_data)

    # Step 2: Calculate the Z-statistic
    z_statistic = (sample_mean - population_mean) / (population_std / np.sqrt(n))

    # Step 3: Calculate the p-value for the two-tailed test
    p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))

    # Step 4: Print the results
    print(f"Sample Mean: {sample_mean}")
    print(f"Z-statistic: {z_statistic}")
    print(f"P-value: {p_value}")

    # Step 5: Hypothesis testing decision
    if p_value < alpha:
        print(f"Reject the null hypothesis: The sample mean is significantly different from {population_mean}.")
    else:
        print(f"Fail to reject the null hypothesis: The sample mean is not significantly different from {population_mean}.")

# Example sample data
sample_data = [52, 55, 49, 48, 50, 53, 51, 54, 52, 50]

# Known population parameters
population_mean = 50      # Null hypothesis mean
population_std = 5        # Population standard deviation

# Perform hypothesis testing
z_test(sample_data, population_mean, population_std, alpha=0.05)
```

```python
## Create a confidence interval for a dataset using Python and interpret the result
import numpy as np
import scipy.stats as stats

def calculate_confidence_interval(sample_data, confidence_level=0.95):
    """
    Calculate the confidence interval for the mean of a sample dataset.

    Parameters:
    sample_data (array-like): The sample dataset.
    confidence_level (float): The confidence level (default is 0.95 for 95% confidence).

    Returns:
    tuple: A tuple containing the lower and upper bounds of the confidence interval.
    """
    # Step 1: Calculate the sample mean and standard deviation
    sample_mean = np.mean(sample_data)
    sample_std_dev = np.std(sample_data, ddof=1)  # Use ddof=1 for sample standard deviation
    n = len(sample_data)

    # Step 2: Calculate the standard error of the mean
    SE = sample_std_dev / np.sqrt(n)

    # Step 3: Calculate the t-critical value for the given confidence level
    alpha = 1 - confidence_level
    t_critical = stats.t.ppf(1 - alpha / 2, df=n - 1)

    # Step 4: Calculate the margin of error
    margin_of_error = t_critical * SE

    # Step 5: Calculate the confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound

# Example sample data
sample_data = [52, 55, 49, 48, 50, 53, 51, 54, 52, 50]

# Calculate the confidence interval
lower_bound, upper_bound = calculate_confidence_interval(sample_data, confidence_level=0.95)

# Print the results
print(f"Sample Data: {sample_data}")
print(f"95% Confidence Interval: ({lower_bound}, {upper_bound})")
```

```python
## Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean.
import numpy as np
import scipy.stats as stats

def calculate_confidence_interval(sample_data, confidence_level=0.95):
    """
    Calculate the confidence interval for the mean of a sample dataset.

    Parameters:
    sample_data (array-like): The sample dataset.
    confidence_level (float): The confidence level (default is 0.95 for 95% confidence).

    Returns:
    tuple: A tuple containing the lower and upper bounds of the confidence interval.
    """
    # Step 1: Calculate the sample mean and standard deviation
    sample_mean = np.mean(sample_data)
    sample_std_dev = np.std(sample_data, ddof=1)  # Use ddof=1 for sample standard deviation
    n = len(sample_data)

    # Step 2: Calculate the standard error of the mean
    SE = sample_std_dev / np.sqrt(n)

    # Step 3: Calculate the t-critical value for the given confidence level
    alpha = 1 - confidence_level
    t_critical = stats.t.ppf(1 - alpha / 2, df=n - 1)

    # Step 4: Calculate the margin of error
    margin_of_error = t_critical * SE

    # Step 5: Calculate the confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound

# Step 1: Generate data from a normal distribution
np.random.seed(42)  # For reproducibility
population_mean = 50  # The true mean of the distribution
population_std_dev = 10  # The standard deviation of the distribution
sample_size = 100  # Number of data points to generate

# Generate sample data
sample_data = np.random.normal(loc=population_mean, scale=population_std_dev, size=sample_size)

# Step 2: Calculate the confidence interval for the sample data
lower_bound, upper_bound = calculate_confidence_interval(sample_data, confidence_level=0.95)

# Step 3: Print the results
print(f"Sample Data (First 10 values): {sample_data[:10]}")
print(f"Sample Mean: {np.mean(sample_data)}")
print(f"95% Confidence Interval: ({lower_bound}, {upper_bound})")
```

```python
## Write a Python script to calculate and visualize the probability density function (PDF) of a normal distribution
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Define the parameters of the normal distribution
mean = 0  # mean of the distribution
std_dev = 1  # standard deviation of the distribution

# Step 1: Generate a range of values (x) from -4 to 4 (for visualization purposes)
x_values = np.linspace(-4, 4, 1000)

# Step 2: Calculate the PDF for each x value
pdf_values = stats.norm.pdf(x_values, loc=mean, scale=std_dev)

# Step 3: Visualize the PDF
plt.figure(figsize=(8, 6))
plt.plot(x_values, pdf_values, label=f'Normal Distribution\n$\mu={mean}, \sigma={std_dev}$', color='b')
plt.fill_between(x_values, pdf_values, alpha=0.2, color='blue')  # Optional: Fill the area under the curve
plt.title('Probability Density Function (PDF) of a Normal Distribution')
plt.xlabel('X values')
plt.ylabel('Probability Density')
plt.grid(True)
plt.legend()
plt.show()
```

```python
##  Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Define the parameter for the Poisson distribution
lambda_ = 3  # Mean (and variance) of the distribution

# Step 1: Generate a range of values (x) for which to calculate the CDF
x_values = np.arange(0, 10)  # We will calculate the CDF for values from 0 to 9

# Step 2: Calculate the CDF for each x value using the Poisson distribution
cdf_values = stats.poisson.cdf(x_values, mu=lambda_)

# Step 3: Plot the CDF
plt.figure(figsize=(8, 6))
plt.plot(x_values, cdf_values, marker='o', color='b', label=f'Poisson CDF\n$\lambda={lambda_}$')
plt.title('Cumulative Distribution Function (CDF) of a Poisson Distribution')
plt.xlabel('X values')
plt.ylabel('Cumulative Probability')
plt.grid(True)
plt.legend()
plt.show()

# Print CDF values for interpretation
for x, cdf in zip(x_values, cdf_values):
    print(f"CDF at X = {x}: {cdf:.4f}")
```

```python
## Simulate a random variable using a continuous uniform distribution and calculate its expected value
import numpy as np

# Define the parameters of the continuous uniform distribution
a = 0  # Lower bound
b = 10  # Upper bound

# Step 1: Simulate a random variable using the continuous uniform distribution
num_samples = 1000  # Number of random samples to simulate
random_samples = np.random.uniform(a, b, num_samples)

# Step 2: Calculate the expected value (theoretical)
expected_value = (a + b) / 2

# Step 3: Calculate the sample mean
sample_mean = np.mean(random_samples)

# Output the results
print(f"Theoretical Expected Value: {expected_value}")
print(f"Sample Mean (from simulation): {sample_mean}")

# Step 4: Visualize the distribution of the random variable
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.hist(random_samples, bins=30, density=True, color='skyblue', alpha=0.7)
plt.title('Continuous Uniform Distribution - Random Variable Simulation')
plt.xlabel('Value')
plt.ylabel('Density')
plt.axvline(expected_value, color='r', linestyle='dashed', linewidth=2, label=f'Expected Value: {expected_value}')
plt.legend()
plt.grid(True)
plt.show()
```

```python
##  Implement Z-score normalization on a dataset and visualize its transformation
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate a dataset from a normal distribution
np.random.seed(42)
data = np.random.normal(loc=100, scale=20, size=1000)  # Mean=100, SD=20

# Step 2: Z-score normalization
mean = np.mean(data)
std = np.std(data)
z_scores = (data - mean) / std

# Step 3: Verify the new mean and standard deviation
normalized_mean = np.mean(z_scores)
normalized_std = np.std(z_scores)

print(f"Original Mean: {mean:.2f}, Original Std Dev: {std:.2f}")
print(f"Z-score Normalized Mean: {normalized_mean:.2f}, Std Dev: {normalized_std:.2f}")

# Step 4: Visualization
plt.figure(figsize=(12, 5))

# Original data
plt.subplot(1, 2, 1)
plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.title("Original Data Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(True)

# Normalized data
plt.subplot(1, 2, 2)
plt.hist(z_scores, bins=30, color='salmon', edgecolor='black')
plt.title("Z-score Normalized Data")
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
## Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.
import numpy as np
from scipy.stats import skew, kurtosis

def calculate_skewness_kurtosis(data):
    """
    Calculate the skewness and kurtosis of a dataset.

    Parameters:
    - data: array-like, the input dataset.

    Returns:
    - tuple: (skewness, kurtosis)
    """
    # Calculate skewness
    skew_value = skew(data)

    # Calculate kurtosis (Fisher's definition, excess kurtosis)
    kurt_value = kurtosis(data, fisher=True)

    return skew_value, kurt_value

# Example usage:
np.random.seed(42)
data = np.random.normal(loc=0, scale=1, size=1000)

skewness_value, kurtosis_value = calculate_skewness_kurtosis(data)
print(f"Skewness: {skewness_value:.4f}")
print(f"Kurtosis: {kurtosis_value:.4f}")
```
